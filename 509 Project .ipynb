```python
fig, ax = plt.subplots(1, 3, figsize=(18,4))

fraud = data['Class'].values
amount = data['Amount'].values
time = data['Time'].values

sns.distplot(fraud, ax=ax[0], color='g')
ax[0].set_title('Distribution of Fraud Amount', fontsize=14)
ax[0].set_xlim([min(fraud), max(fraud)])

sns.distplot(amount, ax=ax[1], color='r')
ax[1].set_title('Distribution of Transaction Amount', fontsize=14)
ax[1].set_xlim([min(amount), max(amount)])

sns.distplot(time, ax=ax[2], color='b')
ax[2].set_title('Distribution of Transaction Time', fontsize=14)
ax[2].set_xlim([min(time), max(time)])
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
rob_scaler = RobustScaler()
```

```python
data['scaled_amount'] = rob_scaler.fit_transform(data['Amount'].values.reshape(-1,1))
data['scaled_time'] = rob_scaler.fit_transform(data['Time'].values.reshape(-1,1))
```

```python
data.drop(['Time','Amount'], axis=1, inplace=True)
```

```python
data = data.sample(frac=1)

fraud_df = data.loc[data['Class'] == 1]
non_fraud_df = data.loc[data['Class'] == 0][:len(fraud_df)]

normal_distributed_data = pd.concat([fraud_df, non_fraud_df])

balanced_data = normal_distributed_data.sample(frac=1, random_state=42)

balanced_data.head()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(18,4))

fraud = balanced_data['Class'].values
amount = balanced_data['scaled_amount'].values
time = balanced_data['scaled_time'].values

sns.distplot(fraud, ax=ax[0], color='g')
ax[0].set_title('Distribution of Fraud Amount', fontsize=14)
ax[0].set_xlim([min(fraud), max(fraud)])

sns.distplot(amount, ax=ax[1], color='r')
ax[1].set_title('Distribution of Transaction Amount', fontsize=14)
ax[1].set_xlim([min(amount), max(amount)])

sns.distplot(time, ax=ax[2], color='b')
ax[2].set_title('Distribution of Transaction Time', fontsize=14)
ax[2].set_xlim([min(time), max(time)])
```

```python
print('Not fraudulent', round(balanced_data['Class'].value_counts()[0]/len(balanced_data) * 100,2), '% of the dataset')
print('Fraudulent', round(balanced_data['Class'].value_counts()[1]/len(balanced_data) * 100,2), '% of the dataset')
```

```python
sample_corr = balanced_data.corr()
fig, ax = plt.subplots(figsize=(20,10))  
sns.heatmap(sample_corr, cmap='coolwarm_r', ax=ax)
plt.show()
```
